## Creating InputExample 

```python
from sentence_transformers import InputExample
from tqdm.auto import tqdm


train = []
for split in tqdm(squad_bn):
  for row in tqdm(split):
    train.append(InputExample(
      texts = [row["question"], row["context"]]
  ))
```

### adding custom data files to `InputExample`

```python
import pandas as pd
df = pd.read_excel("/kaggle/input/mujib-context/context_n_question.xlsx")
df.head()
```

```python
df.isna().sum()
```

```python
df = df[['CONTEXT','QUESTION']]
df.isna().sum()
```

```python
df = df.dropna()
df.isna().sum()
```

```python
for _, row in tqdm(df.iterrows()):
    train.append(InputExample(texts=[row["QUESTION"], row["CONTEXT"]]))
```

```python
train_len = len(train)
train_len
```

### Omitting Duplicate Values

```python
from sentence_transformers import datasets
batch_size = 8

loader = datasets.NoDuplicatesDataLoader(
    train, batch_size = batch_size
)
```

```python
for data in loader :
    print(f"The len of each loader : {len(data)}")
    print("="*60)
    for d in data: 
        print(d.texts)
    break
```

```python
train_no_dups_len = len(loader)*batch_size
train_no_dups_len
```

```python
duplicates = train_len - train_no_dups_len
duplicates
```

```python
model_location = "/kaggle/input/fork-of-dpr-mpnet-tuned-6lakh-epoch24/TUNED_DPR_6LAKH_Epoch_26"
```

```python
from sentence_transformers import  SentenceTransformer
model = SentenceTransformer(model_location)
model.to('cuda:1')
```

### Training

```python
from sentence_transformers import losses

loss = losses.MultipleNegativesRankingLoss(model)
```

```python
!nvidia-smi
```

```python
import os
out_dir = "mujijb_dpr_v5"
if not os.path.exists(out_dir):
    os.mkdir(out_dir)
```

```python
epochs = 30
warmup_steps = int(len(loader) * epochs * 0.1)

model.fit(
    train_objectives=[(loader, loss)],
    epochs=epochs,
    warmup_steps=warmup_steps,
    output_path=out_dir,
#     optimizer_params={'lr': 2e-5, 'eps': 1e-6},#'correct_bias':False}, 
    save_best_model=True,
    checkpoint_path = out_dir+'/checkpoint',
    checkpoint_save_steps = 1500,
    checkpoint_save_total_limit = 2,
    show_progress_bar=True
)
```
